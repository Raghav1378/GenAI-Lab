# Simple LLM calling with streaming

```python
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage,SystemMessage
```

```python
model=init_chat_model("groq:llama-3.1-8b-instant",streaming=True)
```

```python
messages = [
    SystemMessage("You are a helpful AI assistant"),
    HumanMessage("What is the use of the Langchain Library?")
]
```

```python
for chunks in model.stream(messages):
    print(chunks.content,end="",flush=True)
```
