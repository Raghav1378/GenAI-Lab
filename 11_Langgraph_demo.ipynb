```python
from serpapi import GoogleSearch
from dotenv import load_dotenv
load_dotenv()
import os 
serp_api=os.getenv("SERPAPI_API_KEY")
params = {
  "engine": "google",
  "q": "Is coffee healthy",
  "api_key": serp_api
}

search = GoogleSearch(params)
results = search.get_dict()
trusted_sites = [
    "bbc.com", "reuters.com", "who.int", "nature.com", "nytimes.com",
    "theguardian.com", "mayoclinic.org", "healthline.com",
    "medicalnewstoday.com", "webmd.com"
]

for r in results.get("organic_results", []):
    title = r.get("title", "")
    link = r.get("link", "")
    if any(site in link for site in trusted_sites):
        print("✅ Credible:", title, "-", link)
    else:
        print("⚠️ Check yourself:", title, "-", link)

# pick the normal search results
```
