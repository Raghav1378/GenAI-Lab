```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
load_dotenv()
import os
```

```python
load_dotenv()
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant. Please respond to the user request only based on the given context."),
        ("user", "Question: {question}\nContext: {context}")
    ]
)
```

```python
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
output_parser=StrOutputParser()

chain =  prompt |  model | output_parser
```

```python
question = "Can you summarize this story?"
context="""In a quiet corner of the old city stood a shop filled with ticking hearts — clocks of every size, each one whispering time in its own rhythm. The clockmaker, Mr. Arin, had built them all by hand. He never married, never traveled, and never missed a sunrise.

One day, a small boy named Dev came into the shop carrying a broken pocket watch. It had belonged to his late mother. “Can you fix it?” he asked.

Arin studied the watch carefully. Its gears were rusted, the glass cracked, and the spring half-snapped. “It will take time,” he said softly.

Dev visited every evening, sitting by the counter, asking questions about the stars, the city, and how time worked. Slowly, the clockmaker began to smile again — something he hadn’t done in years.

After many nights, the watch finally ticked again. But before Dev could return to collect it, Arin’s heart stopped. When the boy arrived the next morning, the shop was silent.

On the counter lay the mended watch and a note:

“Time cannot be repaired, only remembered. Keep it running — for both of us.”

Dev wound the watch, and as its tiny hands began to move, he swore he could hear two heartbeats ticking in sync — one made of gears, one of love."""
```

```python
print(chain.invoke({"question":question,"context":context}))
```
