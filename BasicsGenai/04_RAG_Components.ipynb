# Vector Stores(Vector database)

## Store embeddinngs in very fast searchbale database

```python
from langchain_community.document_loaders import TextLoader
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
```

```python
loaded_document=TextLoader('rag.txt').load()
```

```python
text_splitter2=CharacterTextSplitter(chunk_size=1000,chunk_overlap=0)
chunks_of_text=text_splitter2.split_documents(loaded_document)
```

```python
vector_db=Chroma.from_documents(chunks_of_text,embedding=embedding_model)
```

```python
Question="Why is RAG better than a regular LLM?"

response=vector_db.similarity_search(Question)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
chat_llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")  # or "gemini-2.0" if available
```

```python
from langchain.chains import RetrievalQA
qa = RetrievalQA.from_chain_type(
    llm=chat_llm,
    retriever=vector_db.as_retriever(),
    chain_type="stuff"
)
```

```python
question = "What are the advantages of the Rag in short,List some use cases of RAG but in detail.,How can FAISS or Chroma be used in a RAG pipeline? in very detail"
response = qa.run(question)
```
